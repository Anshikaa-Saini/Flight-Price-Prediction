```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing training data
train_df=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Data_Train.xlsx")
train_df.head(10)
```

```python
#importing test data
test_df=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Test_set.xlsx")
test_df.head(10)
```

```python

final_df=pd.concat([train_df, test_df], ignore_index=True)
final_df.head(10)
```

```python
final_df.info()
```

```python
#Feature Engineering
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(2)
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
final_df.head(2)
```

```python
final_arrival_time=final_df['Arrival_Time'].str.split(' ').str[0]
final_arrival_time.head(2)
#split using lambda fn
#final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_Hour']=final_arrival_time.str.split(':').str[0]
final_df['Arrival_Minutes']=final_arrival_time.str.split(':').str[1]
```

```python
final_df.head(2)
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minutes']=final_df['Arrival_Minutes'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
final_df.head(2)
```

```python
final_df['Dept_Hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_Minutes']=final_df['Dep_Time'].str.split(':').str[1]
final_df['Dept_Hour']=final_df['Dept_Hour'].astype(int)
final_df['Dept_Minutes']=final_df['Dept_Minutes'].astype(int)
final_df.head(2)
```

```python
final_df.drop('Dep_Time', axis=1, inplace=True)
final_df.head(2)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, 'nan':1})
```

```python
final_df.head(2)
```

```python
final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head(2)
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```
